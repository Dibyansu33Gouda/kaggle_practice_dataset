## 1. Import

```python
import pandas as pd
import numpy as np 
import warnings
```

### 2 . Loading the dataset 

loading the file and reading it and making a copy dataframe for safe work .

```python
pd.set_option('display.max_columns',100)
pd.set_option('display.float_format', lambda x:'%3f' % x)
warnings.filterwarnings('ignore')
Raw_file_path=r"D:\fifa_world_cup_2026_player_performance.csv"
raw_df=pd.read_csv(Raw_file_path)
df=raw_df.copy()

print(f"succesfully loaded , working copy loaded to memory")
print(f"Initial Dimension :{df.shape[0]} rows * {df.shape[1]} columns")
```

### 3. Inspection
inspection about memory usage  , datratypes , displaying 3 players details & stastical analysis of the dataset.

```python
print("-------Dataset_structure---------")
df.info(memory_usage ='deep')
display(df.head(3))
print("----statistical_Analysis-------")
display(df.describe().T[['count' ,'mean' , 'std' , 'max' ,'50%']])
```

### 4.Missing Values Audit & Structural Handling
critical errors , null values and structural implimentations are handeled.

```python
missing_count=df.isnull().sum()
missing_per=(missing_count/len(df))*100
missing_audit=pd.DataFrame({
    'Missing_Count':missing_count ,
    'Missing_Percentage' : missing_per,
    'Datatype' : df.dtypes
}).sort_values(by='Missing_Count',ascending=False)
display(missing_audit[missing_audit['Missing_Count']>0])

gk_specific_cols = ['saves', 'save_percentage', 'punches', 'clean_sheet', 'goals_conceded', 'penalty_saves']

outfield_mas=df['position']!='Goalkeeper'
for col in gk_specific_cols:
    if col in df.columns:
        df.loc[outfield_mas , col]=df.loc[outfield_mas , col].fillna(0.0)
outfield_specific_cols = ['crosses', 'successful_crosses', 'shots_on_target']
gk_mask = df['position'] == 'Goalkeeper'
for col in outfield_specific_cols:
    if col in df.columns:
        df.loc[gk_mask, col] = df.loc[gk_mask, col].fillna(0.0)
# Sanity: ensure position column wasn't accidentally overwritten earlier
assert df['position'].nunique() == 4, 'Unexpected number of unique positions after imputation'
critical_cols=['player_id' , 'match_id' , 'minutes_played' , 'goals']
assert df[critical_cols].isnull().sum().sum()==0,"CRITICAL ERROR: Unexpected NaNs in primary key or core metric columns!"
print("structural implimentation complete  .  domain specific nulls handeled carefully")
```

### 5. Duplicate check 
checking duplicates and removing them if found

```python
composite_key=['player_id' , 'match_id']
duplicate_mask=df.duplicated(subset=composite_key , keep=False)
duplicate_count=duplicate_mask.sum()

print(f"total duplicate rows found on composite key {composite_key}:{duplicate_count}")
if duplicate_count > 0:
    print("previewing duplicate counts")
    display(df[duplicate_mask].sort_values(by=composite_key).head())
    
    initial_keys= len(df)
    df=df.drop_duplicates(subset=composite_key , keep='first').reset_index(drop=True)
    print(f"Action taked dropped {initial_keys -  len(df)} dup;icate records based on {composite_key}")
else:
    print(f"no dup;icate keys found , composite key uniqueness confirmed ")
```

### 6.Datatype fixes
fixed datatypes of dates and managed memory savings

```python
df['match_date'] = pd.to_datetime(df['match_date'], format=r'%Y-%m-%d', errors='coerce')
assert df['match_date'].isnull().sum() == 0, 'Some match_date values could not be parsed'
catagoreal_columns=[
    'position',
    'preferred_foot',
    'tournament_stage',
    'match_result',
    'nationality',
    'team'
    
]
for col in catagoreal_columns:
    if col in df.columns:
        df[col]=df[col].astype('category')
int_columns=['goals', 'assists', 'shots', 'key_passes', 'yellow_cards', 'red_cards', 'minutes_played']
for col in int_columns:
    if col in df.columns:
        df[col]=df[col].fillna(0).astype('Int64')
print("------Updated datatypes and memory management")
df.info(memory_usage='deep')
```

### 7.Derived Columns

```python
# 1. Expected Goals & Assists Differentials (Finishing / Playmaking Efficiency)
df['xg_diff'] = df['goals'] - df['expected_goals_xg']
df['xa_diff'] = df['assists'] - df['expected_assists_xa']

# 2. Direct Goal Contributions per Match
df['goal_contributions'] = df['goals'] + df['assists']

# 3. Pass Efficiency Check (Safely compute ratio avoiding division by zero)
df['calc_pass_accuracy'] = np.where(
    df['total_passes'] > 0, 
    (df['successful_passes'] / df['total_passes']) * 100, 
    0.0
)

# 4. Action Intensity per 90 Minutes (Standardizing by playing time)
df['actions_per_90'] = np.where(
    df['minutes_played'] > 0,
    ((df['total_passes'] + df['tackles'] + df['interceptions']) / df['minutes_played']) * 90,
    0.0
)

print(f"Feature Engineering complete. Added derived features: ['xg_diff', 'xa_diff', 'goal_contributions', 'calc_pass_accuracy', 'actions_per_90']")
display(df[['player_name', 'goals', 'expected_goals_xg', 'xg_diff', 'actions_per_90']].head(5))
```

### 8.Sanity Checks & Data Integrity Audits

```python
# Sanity Check: Verify if cumulative match-level goals equal the reported `total_goals_tournament`
# Group by player and sum their individual match goals
player_match_goals = df.groupby('player_id')['goals'].sum().reset_index()
player_match_goals.rename(columns={'goals': 'computed_total_goals'}, inplace=True)

# Merge back with the tournament metadata level reported in the records
integrity_check = df[['player_id', 'player_name', 'total_goals_tournament']].drop_duplicates()
integrity_check = integrity_check.merge(player_match_goals, on='player_id', how='left')

# Calculate discrepancy
integrity_check['discrepancy'] = integrity_check['total_goals_tournament'] - integrity_check['computed_total_goals']
mismatches = integrity_check[integrity_check['discrepancy'] != 0]

print("--- SANITY CHECK RESULTS ---")
if len(mismatches) == 0:
    print("SUCCESS: 100% Data Integrity confirmed. Aggregated match goals precisely equal 'total_goals_tournament' across all players.")
else:
    print(f"WARNING: Detected {len(mismatches)} players with mismatching tournament goal aggregations!")
    display(mismatches.head(10))
    # Note: If discrepancies exist, trust granular match-level aggregations (`computed_total_goals`) over static metadata.
```

### 9.Export Cleaned Dataset

```python
CLEANED_OUTPUT_PATH = "cleaned_player_data.csv"

# Export without DataFrame index to prevent unnamed index columns upon reload
df.to_csv(CLEANED_OUTPUT_PATH, index=False)

print(f"Pipeline Execution Complete.")
print(f"Cleaned dataset exported to: '{CLEANED_OUTPUT_PATH}'")
print(f"Final Export Shape: {df.shape[0]:,} rows × {df.shape[1]} columns.")
```

* **Core Identifiers:** Verified `0` missing values across composite primary keys (`player_id`, `match_id`) and core continuous metrics.

* **Temporal Cast:** Converted `match_date` from string representation (`object`) to standard `datetime64[ns]` for chronological sorting and time-series aggregation.
* **Memory Reduction:** Cast low-cardinality discrete strings (`position`, `preferred_foot`, `tournament_stage`, `match_result`, `team`) to `category` dtypes, drastically reducing memory consumption while optimizing groupby performance.
* **Discrete Standardization:** Forced integer casting (`int64`) on count metrics (`goals`, `assists`, `minutes_played`, `cards`).

### **4. Engineered Features for Notebook 2**
* Built **`xg_diff`** (`goals - expected_goals_xg`) and **`xa_diff`** (`assists - expected_assists_xa`) to isolate clinical finishing and playmaking over/under-performance.
* Built **`actions_per_90`** to normalize volume metrics across players with variable substitution minutes.

### **5. Sanity Check Audit**
* Cross-verified granular row-level data against summary metadata. Aggregated match-level `goals` matched `total_goals_tournament` across the player roster with **0 discrepancies**, validating raw dataset collection integrity.
