****⚽ Notebook 2: Player & Team Aggregations (Groupby Phase)****

***Objective: Perform robust, statistically sound aggregations at the Player, Position, and Team levels.***

### 1.Import & load clean dataset

```python
import pandas as pd
import numpy as np
import warnings
pd.set_option('display.max_column', 50)

df = pd.read_csv(r"D:\DS Python\pandaass\cleaned_player_data.csv")
print(f"cleaned data landed: {df.shape[0]} rows * {df.shape[1]} columns")
```

```python
player_agg =(
    df.groupby(['player_id', 'player_name', 'team'])
    .agg(
          total_matches_played=('match_id', 'count'),
          sum_minutes=('minutes_played', 'sum'),
          sum_goals=('goals', 'sum'),
          sum_assist=('assists', 'sum'),
          sum_successful_pass=('successful_passes', 'sum'),
          meta_total_goals=('total_goals_tournament', 'first')
      )

      .reset_index()
)
player_agg.head()
player_agg['goal_match_status']=player_agg['sum_goals']==player_agg['meta_total_goals']
discrepancies=player_agg[~player_agg['goal_match_status']]
if len(discrepancies)==0:
    print("integrity chcek passed : match level aggregrated goal match totalgoal  turnment matchdata")
else:
    print(f"⚠️ WARNING: Found {len(discrepancies)} discrepancies between aggregated match goals and tournament metadata.")
    display(discrepancies[['player_name', 'sum_goals', 'meta_total_goals', 'total_matches_played']])
display(player_agg.sort_values(by='sum_goals', ascending=False).head())
```

```python
print("--- 2. POSITION-LEVEL AGGREGATION (AVOIDING STATISTICAL TRAPS) ---")

# Step A: Create weighted numerators for continuous metrics (weighting by minutes played)
df['weighted_rating'] = df['player_rating'] * df['minutes_played']
df['weighted_distance'] = df['distance_covered_km'] * df['minutes_played']
df['weighted_defense'] = df['defensive_contribution'] * df['minutes_played']

# Step B: Group by position and sum up the bases
position_agg = df.groupby('position').agg(
    total_minutes=('minutes_played', 'sum'),
    total_successful_passes=('successful_passes', 'sum'),
    total_passes_attempted=('total_passes', 'sum'),
    sum_weighted_rating=('weighted_rating', 'sum'),
    sum_weighted_distance=('weighted_distance', 'sum'),
    sum_weighted_defense=('weighted_defense', 'sum')
).reset_index()

# Step C: Calculate the final True Ratios and Weighted Averages
position_agg['true_pass_accuracy_%'] = (position_agg['total_successful_passes'] / position_agg['total_passes_attempted']) * 100

# For continuous metrics, divide the weighted sum by total minutes to get the true average
position_agg['avg_rating_weighted'] = position_agg['sum_weighted_rating'] / position_agg['total_minutes']

# Distance covered is usually evaluated "per 90 minutes" rather than per minute
position_agg['avg_distance_per_90'] = (position_agg['sum_weighted_distance'] / position_agg['total_minutes']) * 90
position_agg['avg_def_contribution'] = position_agg['sum_weighted_defense'] / position_agg['total_minutes']

# Drop the raw calculation columns for a clean final table
cols_to_display = ['position', 'total_minutes', 'true_pass_accuracy_%', 'avg_rating_weighted', 'avg_distance_per_90', 'avg_def_contribution']
display(position_agg[cols_to_display].sort_values(by='avg_rating_weighted', ascending=False).round(2))
```
